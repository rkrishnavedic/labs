```
import math
import matplotlib.pyplot as plt
import numpy as np

#Problem 1
print('Problem 1');

def f(t):
    return math.exp(t);

def firstOrder(a,b,x):
    return ((x-b)/(a-b))*f(a) + ((x-a)/(b-a))*f(b);

def secondOrder(a,b,c,x):
    return (((x-c)*(x-b))/((a-c)*(a-b)))*f(a)+(((x-c)*(x-a))/((b-a)*(b-c)))*f(b)+(((x-a)*(x-b))/((c-a)*(c-b)))*f(c);

print('\n Part (i):');
approx = firstOrder(0,0.5,0.25);
exact = f(0.25);
print('For x0 = 0 and x1 = 0.5, we have approximation = '+str(approx)+' while f(0.25) = '+str(exact));
print('Absolute Error = '+str(abs(exact-approx)));

print('\n Part (ii):');
approx = firstOrder(0.5,1,0.75);
exact = f(0.75);
print('For x0 = 0.5 and x1 = 1, we have approximation = '+str(approx)+' while f(0.75) = '+str(exact));
print('Absolute Error = '+str(abs(exact-approx)));

print('\n Part (iii):');
approx = secondOrder(0,1,2,0.25);
exact = f(0.25);
print('For x0 = 0, x1=1 and x2 = 2, we have approximation = '+str(approx)+' while f(0.25) = '+str(exact));
print('Absolute Error = '+str(abs(exact-approx)));

approx = secondOrder(0,1,2,0.75);
exact = f(0.75);
print('For x0 = 0, x1=1 and x2 = 2, we have approximation = '+str(approx)+' while f(0.75) = '+str(exact));
print('Absolute Error = '+str(abs(exact-approx)));

#let us plot to see

X = np.linspace(0,1,1000);
Y1 = [0]*1000;
Y2 = [0]*1000;
f1 = np.vectorize(f);
for i in range(1000):
    Y1[i] = firstOrder(0,0.5,X[i]);
    Y2[i] = secondOrder(0,1,2,X[i]);

plt.plot(X,Y1, label='$1^{st}$ order');
plt.plot(X,Y2, label='$2^{nd}$ order');
plt.plot(X, f1(X), label ='f(x)' );
plt.legend();
plt.show();

Z1 = [0]*1000;
Z2 = [0]*1000;

for i in range(1000):
    Z1[i] = firstOrder(0.5,1.0,X[i]);
    Z2[i] = secondOrder(0,1,2,X[i]);

plt.plot(X,Z1, label='$1^{st}$ order');
plt.plot(X,Z2, label='$2^{nd}$ order');
plt.plot(X, f1(X), label ='f(x)' );
plt.legend();

plt.show();

```

```
import math
import matplotlib.pyplot as plt
import numpy as np

#Problem 4
print('Problem 4');

def f(t):
    return math.log10(math.tan(t));

def f1(t):
    if t==1.0:
        return 0.1924;
    if t==1.05:
        return 0.2414;
    if t==1.10:
        return 0.2933;
    if t==1.15:
        return 0.3492;

def thirdOrder1(a,b,c,d,x):
    return (((x-b)*(x-c)*(x-d))/((a-b)*(a-c)*(a-d)))*f1(a) +(((x-a)*(x-c)*(x-d))/((b-a)*(b-c)*(b-d)))*f1(b) +(((x-b)*(x-a)*(x-d))/((c-b)*(c-a)*(c-d)))*f1(c) +(((x-b)*(x-c)*(x-a))/((d-b)*(d-c)*(d-a)))*f1(d);

print('approximation of f(1.09) = '+str(round(thirdOrder1(1.0,1.05,1.10,1.15,1.09),4)));

X = np.linspace(0.5,1.5,1000);
Y = [0]*1000;
Z = [0]*1000;
max_error = 0;
avg_error = 0;
for i in range(1000):
    Y[i] = f(X[i]);
    Z[i] = thirdOrder1(1.0,1.05,1.10,1.15,X[i]);
    max_error = max(max_error, abs(Y[i]-Z[i]));
    avg_error += abs(Y[i]-Z[i]);

plt.plot(X,Y, label='true function');
plt.plot(X,Z,label = 'approximation function');
plt.legend();
plt.show();

print('error in approximation of f(1.09) = '+str(round(abs(thirdOrder1(1.0,1.05,1.10,1.15,1.09)-f(1.09)))))
print('max error as bound in the interval [0.5,1.5] = '+str(round(max_error,4)));
print('average absolute error in the interval [0.5,1.5] = '+str(round(avg_error/1000, 4)));
```
