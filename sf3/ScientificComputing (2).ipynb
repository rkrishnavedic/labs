```
import math
import matplotlib.pyplot as plt
import numpy as np
 
print('Problem 1');
 
def newtonForwardCoeff(y):
    n = len(y);
    n-=1;
    #assuming x_i - x_{i-1} = h = constant;
    a = np.copy(y);
    for i in range(n+1):
        j=n;
        while j>i:
            a[j]-=a[j-1];
            j-=1;
    
    return a;
    
def newtonForward(a,x,t):
    ans = 0;
    n = len(a)-1;
    f = 1;
    p = 1;
    h = x[1]-x[0];
    ph=1;
    for i in range(n+1):
        ans+=(a[i]*p)/(f*ph);
        p*=(t-x[i]);
        ph*=h;
        f*=i+1;
    return ans;
 
 
print('\npart (i) :')
N = 100;
x1 = np.linspace(0,0.75,N);
y1 = [0]*N;
y2 = [0]*N;
y3 = [0]*N;
 
X = [0.25,0.5];
Y = [1.64872,  2.71828];
A = newtonForwardCoeff(Y);
for i in range(N):
    y1[i] = newtonForward(A,X,x1[i]);
 
print('one Degree approximation f(0.43) = '+str(newtonForward(A,X,0.43)));
 
X = [0.25,0.5,0.75];
Y = [1.64872,  2.71828, 4.48169];
A = newtonForwardCoeff(Y);
for i in range(N):
    y2[i] = newtonForward(A,X,x1[i]);
print('two Degree approximation f(0.43) = '+str(newtonForward(A,X,0.43)));
 
X = [0.0,0.25,0.5,0.75];
Y = [1.0, 1.64872,  2.71828, 4.48169];
A = newtonForwardCoeff(Y);
for i in range(N):
    y3[i] = newtonForward(A,X,x1[i]);
print('three Degree approximation f(0.43) = '+str(newtonForward(A,X,0.43)));
 
print('Approximation plot:');
 
plt.plot(x1,y1, label ='one degree');
plt.plot(x1,y2, label ='two degree');
plt.plot(x1,y3, label ='third degree');
plt.legend();
plt.xlabel('x');
plt.ylabel('f(x)')
plt.show();
 
 
 
print('\npart (ii) :')
N = 100;
x1 = np.linspace(0.0,0.5,N);
y1 = [0]*N;
y2 = [0]*N;
y3 = [0]*N;
 
X = [0.1,0.2];
Y = [-0.29004986, -0.56079734];
A = newtonForwardCoeff(Y);
for i in range(N):
    y1[i] = newtonForward(A,X,x1[i]);
 
print('one Degree approximation f(0.18) = '+str(newtonForward(A,X,0.18)));
 
X = [0.1,0.2,0.3];
Y = [-0.29004986, -0.56079734,  -0.81401972];
A = newtonForwardCoeff(Y);
for i in range(N):
    y2[i] = newtonForward(A,X,x1[i]);
print('two Degree approximation f(0.18) = '+str(newtonForward(A,X,0.18)));
 
X = [0.1,0.2,0.3,0.4];
Y = [-0.29004986, -0.56079734,  -0.81401972, -1.0526302];
A = newtonForwardCoeff(Y);
for i in range(N):
    y3[i] = newtonForward(A,X,x1[i]);
print('three Degree approximation f(0.18) = '+str(newtonForward(A,X,0.18)));
 
print('Approximation plot:');
 
plt.plot(x1,y1, label ='one degree');
plt.plot(x1,y2, label ='two degree');
plt.plot(x1,y3, label ='third degree');
plt.legend();
plt.xlabel('x');
plt.ylabel('f(x)')
plt.show();
```

```
import math
import matplotlib.pyplot as plt
import numpy as np

print('Problem 5');

def newtonForwardCoeff(y):
    n = len(y);
    n-=1;
    #assuming x_i - x_{i-1} = h = constant;
    a = np.copy(y);
    for i in range(n+1):
        j=n;
        while j>i:
            a[j]-=a[j-1];
            j-=1;
    
    return a;
    
def newtonForward(a,x,t):
    ans = 0;
    n = len(a)-1;
    f = 1;
    p = 1;
    h = x[1]-x[0];
    ph=1;
    for i in range(n+1):
        ans+=(a[i]*p)/(f*ph);
        p*=(t-x[i]);
        ph*=h;
        f*=i+1;
    return ans;

def P(t):
    return 3 - 2*(t+1) + 0*(t+1)*t + t*(t+1)*(t-1);

def Q(t):
    return -1 + 4*(t+2) - 3*(t+2)*(t+1) + (t+2)*(t+1)*t;

X = [-2,-1,0,1,2];
Y = [-1, 3, 1, -1, 3];
A = newtonForwardCoeff(Y);
N =100;
x1 = np.linspace(-3,3,N);
y1 = [0]*N;
y2 = [0]*N;
y3 = [0]*N;
for i in range(N):
    y1[i] = newtonForward(A,X,x1[i]);
for i in range(N):
    y2[i] = P(x1[i]);
for i in range(N):
    y3[i] = Q(x1[i]);

plt.plot(x1,y1, label ='Interpolate f(x)');
plt.plot(x1,y2, label ='P(x)');
plt.plot(x1,y3, label ='Q(x)');
plt.legend();
plt.xlabel('x');
plt.ylabel('y')
plt.show();
```
