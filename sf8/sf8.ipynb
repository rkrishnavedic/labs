```
import math
import numpy as np
import matplotlib.pyplot as plt

print('Problem 1:')

def f(x):
    return 1+x-math.cos(x)-(1+math.pi/2)*math.sin(x);

f1 = np.vectorize(f);

def solve(n, opt):
    h = (math.pi/2)/n;
    # Get A
    A = np.zeros((n+1, n+1))
    A[0, 0] = -(2+h**2)
    A[0,1] = 1
    A[n, n-1] = -(2+h**2)
    A[n, n] = 1
    for i in range(1, n):
        A[i, i-1] = 1
        A[i, i] = -(2+h**2)
        A[i, i+1] = 1

    #print(A)

    # Get b
    b = np.zeros(n+1)
    for i in range(1,n):
        b[i]=(i*h)*(h**2);
    b[n] = (n*h)*(h**2)
    #print(b)

    # solve the linear equations
    y = np.linalg.solve(A, b)

    t = np.linspace(0, math.pi/2, n+1)

    tt = np.linspace(0, math.pi/2, 100);

    yy = f1(tt);
    print('y(pi/4) = '+str(f(math.pi/4))+' fd = '+str(y[n//2]))
    plt.figure(figsize=(10,8))
    plt.plot(t, y)
    plt.plot(tt, yy)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()

a=0;
b=math.pi/2;

H = [4, 8, 16, 32, 64];

for h in H:
    print('for h = '+str(h))
    solve(h, 1)


print('----\n h = 1/64')
solve(64, 0);

```

```
import math
import numpy as np
import matplotlib.pyplot as plt

print('Problem 3:')

def f(x):
    return 2*np.exp(x) - x - 1;

f1 = np.vectorize(f);

def solve(n):
    h = (1)/n;
    # Get A
    A = np.zeros((n+1, n+1))
    A[0, 0] = -(2+h**2)
    A[0,1] = 1
    A[n, n-1] = -(2+h**2)
    A[n, n] = 1
    for i in range(1, n):
        A[i, i-1] = 1
        A[i, i] = -(2+h**2)
        A[i, i+1] = 1

    #print(A)

    # Get b
    b = np.zeros(n+1)
    b[0] = 0
    for i in range(1,n):
        b[i]=(h**2);
    b[n] = (h**2) - (np.exp(1)-1)*(1-h*0.5)
    #print(b)

    # solve the linear equations
    y = np.linalg.solve(A, b)

    t = np.linspace(0, 1, n+1)

    tt = np.linspace(0, 1, 100);

    yy = f1(tt);
    print('y(1/2) = '+str(f(0.5))+' fd = '+str(y[n//2]))
    plt.figure(figsize=(10,8))
    plt.plot(t, y)
    plt.plot(tt, yy)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()


solve(3)

```

```
import math
import numpy as np
import matplotlib.pyplot as plt

print('Problem 4:')

def f(x):
    return (np.exp(10*x)-1)/(np.exp(x)-1);

f1 = np.vectorize(f);

def solve(n):
    h = (1)/n;
    # Get A
    A = np.zeros((n+1, n+1))
    A[0, 0] = -(2-10*h**2)
    A[0,1] = 1
    A[n, n-1] = -(2-10*h**2)
    A[n, n] = 1
    for i in range(1, n):
        A[i, i-1] = 1
        A[i, i] = -(2-10*h**2)
        A[i, i+1] = 1

    #print(A)

    # Get b
    b = np.zeros(n+1)
    b[0] = 0
    for i in range(1,n):
        b[i]=0;
    b[n] = -1
    #print(b)

    # solve the linear equations
    y = np.linalg.solve(A, b)

    t = np.linspace(0, 1, n+1)

    tt = np.linspace(0, 1, 10);

    yy = f1(tt);
    plt.figure(figsize=(10,8))
    plt.plot(t, y)
    plt.plot(tt, yy)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()


solve(3)

```
